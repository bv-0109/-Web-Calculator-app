```python
# Install Gradio
!pip install gradio

import gradio as gr

# Function to handle button presses
def press_button(current, btn):
    """
    current: string representing the current display value
    btn: string representing the button pressed
    """
    if btn == "C":  # Clear button resets display
        return ""
    elif btn == "=":  # Equals button evaluates the expression
        try:
            # eval() allows float operations like 3.5 + 2.1
            return str(eval(current))
        except:
            return "Error"
    else:
        # Append the pressed button to the display string
        return current + btn

# Build Gradio interface using Blocks
with gr.Blocks() as demo:
    gr.Markdown("## 🧮 Web Calculator with Buttons")

    # Display textbox (non-editable, shows current input/result)
    display = gr.Textbox(label="Display", interactive=False)

    # Row 1: digits 7-9 and division
    with gr.Row():
        for b in ["7","8","9","/"]:
            gr.Button(b).click(press_button, inputs=[display, gr.State(b)], outputs=display)

    # Row 2: digits 4-6 and multiplication
    with gr.Row():
        for b in ["4","5","6","*"]:
            gr.Button(b).click(press_button, inputs=[display, gr.State(b)], outputs=display)

    # Row 3: digits 1-3 and subtraction
    with gr.Row():
        for b in ["1","2","3","-"]:
            gr.Button(b).click(press_button, inputs=[display, gr.State(b)], outputs=display)

    # Row 4: digit 0, decimal point, equals, addition
    with gr.Row():
        for b in ["0",".","=","+"]:
            gr.Button(b).click(press_button, inputs=[display, gr.State(b)], outputs=display)

    # Row 5: Clear button
    with gr.Row():
        gr.Button("C").click(press_button, inputs=[display, gr.State("C")], outputs=display)

# Launch the calculator app
demo.launch()
```
